# *GSTAR Bobot Seragam*

```python
import pandas as pd
gstar = pd.read_excel('telur_jatim.xlsx')
gstar.head()
```

# **Pembuatan Variabel GSTAR dengan Matriks Pembobot Seragam**

```python
gstar['lag1y1'] = gstar['kediri'].shift(1)
gstar['lag1y2'] = gstar['blitar'].shift(1)
gstar['lag1y3'] = gstar['malang'].shift(1)
gstar['lag2y1'] = gstar['lag1y1'].shift(1)
gstar['lag2y2'] = gstar['lag1y2'].shift(1)
gstar['lag2y3'] = gstar['lag1y3'].shift(1)
gstar['lag3y1'] = gstar['lag2y1'].shift(1)
gstar['lag3y2'] = gstar['lag2y2'].shift(1)
gstar['lag3y3'] = gstar['lag2y3'].shift(1)
```

```python
gstar['difflag1y1'] = (gstar['lag1y1']-gstar['lag2y1'])
gstar['difflag1y2'] = (gstar['lag1y2']-gstar['lag2y2'])
gstar['difflag1y3'] = (gstar['lag1y3']-gstar['lag2y3'])
gstar['difflag2y1'] = (gstar['lag2y1']-gstar['lag3y1'])
gstar['difflag2y2'] = (gstar['lag2y2']-gstar['lag3y2'])
gstar['difflag2y3'] = (gstar['lag2y3']-gstar['lag3y3'])
```

```python
# Calculate the weighted lagged variables
gstar['wlag1y1'] = 0.5 * (gstar['lag1y2'] - gstar['lag2y2']) + 0.5 * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y2'] = 0.5 * (gstar['lag1y1'] - gstar['lag2y1']) + 0.5 * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y3'] = 0.5 * (gstar['lag1y1'] - gstar['lag2y1']) + 0.5 * (gstar['lag1y2'] - gstar['lag2y2'])
gstar['wlag2y1'] = 0.5 * (gstar['lag2y2'] - gstar['lag3y2']) + 0.5 * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y2'] = 0.5 * (gstar['lag2y1'] - gstar['lag3y1']) + 0.5 * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y3'] = 0.5 * (gstar['lag2y1'] - gstar['lag3y1']) + 0.5 * (gstar['lag2y2'] - gstar['lag3y2'])
```

# Train and Test

# *GSTAR Inverse Distance*

```python
import pandas as pd
gstar = pd.read_excel('telur_jatim.xlsx')
gstar.head()
```

# **Pembuatan Matriks Pembobot Invers Jarak**

```python
#jarak kediri ke blitar : 44 km (jarak kediri - blitar)
#jarak kediri ke malang : 100 km (jarak kediri - malang)
#jarak blitar ke malang : 76 km (jarak blitar - malang)
r1 = 44
r2 = 100
r3 = 76
inverse_distance = [[0,r1/(r1+r2),r2/(r1+r2)],
                    [r1/(r1+r3),0,r3/(r1+r3)],
                    [r3/(r2+r3),r2/(r2+r3),0]]
```

```python
inverse_distance
```

# **Pembuatan Variabel GSTAR**

```python
gstar['lag1y1'] = gstar['kediri'].shift(1)
gstar['lag1y2'] = gstar['blitar'].shift(1)
gstar['lag1y3'] = gstar['malang'].shift(1)
gstar['lag2y1'] = gstar['lag1y1'].shift(1)
gstar['lag2y2'] = gstar['lag1y2'].shift(1)
gstar['lag2y3'] = gstar['lag1y3'].shift(1)
gstar['lag3y1'] = gstar['lag2y1'].shift(1)
gstar['lag3y2'] = gstar['lag2y2'].shift(1)
gstar['lag3y3'] = gstar['lag2y3'].shift(1)
```

```python
gstar['difflag1y1'] = (gstar['lag1y1']-gstar['lag2y1'])
gstar['difflag1y2'] = (gstar['lag1y2']-gstar['lag2y2'])
gstar['difflag1y3'] = (gstar['lag1y3']-gstar['lag2y3'])
gstar['difflag2y1'] = (gstar['lag2y1']-gstar['lag3y1'])
gstar['difflag2y2'] = (gstar['lag2y2']-gstar['lag3y2'])
gstar['difflag2y3'] = (gstar['lag2y3']-gstar['lag3y3'])
```

```python
# Calculate the weighted lagged variables
gstar['wlag1y1'] = inverse_distance[0][1] * (gstar['lag1y2'] - gstar['lag2y2']) + inverse_distance[0][2] * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y2'] = inverse_distance[1][0] * (gstar['lag1y1'] - gstar['lag2y1']) + inverse_distance[1][2] * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y3'] = inverse_distance[2][0] * (gstar['lag1y1'] - gstar['lag2y1']) + inverse_distance[2][1] * (gstar['lag1y2'] - gstar['lag2y2'])
gstar['wlag2y1'] = inverse_distance[0][1] * (gstar['lag2y2'] - gstar['lag3y2']) + inverse_distance[0][2] * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y2'] = inverse_distance[1][0] * (gstar['lag2y1'] - gstar['lag3y1']) + inverse_distance[1][2] * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y3'] = inverse_distance[2][0] * (gstar['lag2y1'] - gstar['lag3y1']) + inverse_distance[2][1] * (gstar['lag2y2'] - gstar['lag3y2'])
```

# **Pembuatan Variabel GSTAR**

```python
gstar['lag1y1'] = gstar['kediri'].shift(1)
gstar['lag1y2'] = gstar['blitar'].shift(1)
gstar['lag1y3'] = gstar['malang'].shift(1)
gstar['lag2y1'] = gstar['lag1y1'].shift(1)
gstar['lag2y2'] = gstar['lag1y2'].shift(1)
gstar['lag2y3'] = gstar['lag1y3'].shift(1)
gstar['lag3y1'] = gstar['lag2y1'].shift(1)
gstar['lag3y2'] = gstar['lag2y2'].shift(1)
gstar['lag3y3'] = gstar['lag2y3'].shift(1)
```

```python
gstar['difflag1y1'] = (gstar['lag1y1']-gstar['lag2y1'])
gstar['difflag1y2'] = (gstar['lag1y2']-gstar['lag2y2'])
gstar['difflag1y3'] = (gstar['lag1y3']-gstar['lag2y3'])
gstar['difflag2y1'] = (gstar['lag2y1']-gstar['lag3y1'])
gstar['difflag2y2'] = (gstar['lag2y2']-gstar['lag3y2'])
gstar['difflag2y3'] = (gstar['lag2y3']-gstar['lag3y3'])
```

```python
# Calculate the weighted lagged variables
gstar['wlag1y1'] = norm_cross[0][1] * (gstar['lag1y2'] - gstar['lag2y2']) + norm_cross[0][2] * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y2'] = norm_cross[1][0] * (gstar['lag1y1'] - gstar['lag2y1']) + norm_cross[1][2] * (gstar['lag1y3'] - gstar['lag2y3'])
gstar['wlag1y3'] = norm_cross[2][0] * (gstar['lag1y1'] - gstar['lag2y1']) + norm_cross[2][1] * (gstar['lag1y2'] - gstar['lag2y2'])
gstar['wlag2y1'] = norm_cross[0][1] * (gstar['lag2y2'] - gstar['lag3y2']) + norm_cross[0][2] * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y2'] = norm_cross[1][0] * (gstar['lag2y1'] - gstar['lag3y1']) + norm_cross[1][2] * (gstar['lag2y3'] - gstar['lag3y3'])
gstar['wlag2y3'] = norm_cross[2][0] * (gstar['lag2y1'] - gstar['lag3y1']) + norm_cross[2][1] * (gstar['lag2y2'] - gstar['lag3y2'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming gstarpred['Tanggal'] is a string representing dates in the format 'MM/DD/YYYY'
gstarpred['TANGGAL'] = pd.to_datetime(gstarpred['TANGGAL'], format='%m/%d/%Y')

# Change the date format to '%d/%m/%Y'
gstarpred['TANGGAL'] = gstarpred['TANGGAL'].dt.strftime('%d/%m/%Y')

plt.figure(figsize=(15, 5))  # Adjust the figure size as needed
x = pd.to_datetime(gstarpred['TANGGAL'], format='%d/%m/%Y')

plt.plot(x, gstarpred['KEDIRI_PRED'], label='Kediri (Prediksi)', marker='o', linestyle='dashed', color='#2E4374')
plt.plot(x, gstarpred['BLITAR_PRED'], label='Blitar (Prediksi)', marker='o', linestyle='dashed', color='#B31312')
plt.plot(x, gstarpred['MALANG_PRED'], label='Malang (Prediksi)', marker='o', linestyle='dashed', color='#5D9C59')

plt.xlabel('Tanggal')
plt.ylabel('Harga Telur Ayam Ras (Rupiah)')
plt.title('Prediksi Harga Telur Ayam Ras (Kediri, Blitar, Malang) 5 Hari Kedepan')
# Expand y-axis limits with larger padding
min_y = gstarpred[['KEDIRI', 'KEDIRI_PRED', 'BLITAR', 'BLITAR_PRED', 'MALANG', 'MALANG_PRED']].min().min()
max_y = gstarpred[['KEDIRI', 'KEDIRI_PRED', 'BLITAR', 'BLITAR_PRED', 'MALANG', 'MALANG_PRED']].max().max()
padding = (max_y - min_y) * 0.1  # Add 10% padding
plt.ylim(min_y - padding, max_y + padding)
# Place the legend outside the plot
plt.legend(bbox_to_anchor=(1,1), loc='upper left')

# Set ticks for each day
plt.xticks(x, rotation=0)

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming gstarpred['Tanggal'] is a string representing dates in the format 'MM/DD/YYYY'
gstarpred['TANGGAL'] = pd.to_datetime(gstarpred['TANGGAL'], format='%m/%d/%Y')

# Change the date format to '%d/%m/%Y'
gstarpred['TANGGAL'] = gstarpred['TANGGAL'].dt.strftime('%d/%m/%Y')

plt.figure(figsize=(15, 6))  # Adjust the figure size as needed
x = pd.to_datetime(gstarpred['TANGGAL'], format='%d/%m/%Y')

plt.plot(x, gstarpred['KEDIRI'], label='Kediri (Aktual)', marker='o', linestyle='-', color='#2E4374')
plt.plot(x, gstarpred['KEDIRI_PRED'], label='Kediri (Prediksi)', marker='o', linestyle='dashed', color='#2E4374')
plt.plot(x, gstarpred['BLITAR'], label='Blitar (Aktual)', marker='o', linestyle='-', color='#FFC107')
plt.plot(x, gstarpred['BLITAR_PRED'], label='Blitar (Prediksi)', marker='o', linestyle='dashed', color='#FFC107')
plt.plot(x, gstarpred['MALANG'], label='Malang (Aktual)', marker='o', linestyle='-', color='#388E3C')
plt.plot(x, gstarpred['MALANG_PRED'], label='Malang (Prediksi)', marker='o', linestyle='dashed', color='#388E3C')

plt.xlabel('Tanggal')
plt.ylabel('Harga Telur Ayam Ras(Rupiah)')
plt.title('Prediksi Harga Telur Ayam Ras (Kediri, Blitar, Malang) 5 Hari Kedepan')
# Expand y-axis limits with larger padding
min_y = gstarpred[['KEDIRI', 'KEDIRI_PRED', 'BLITAR', 'BLITAR_PRED', 'MALANG', 'MALANG_PRED']].min().min()
max_y = gstarpred[['KEDIRI', 'KEDIRI_PRED', 'BLITAR', 'BLITAR_PRED', 'MALANG', 'MALANG_PRED']].max().max()
padding = (max_y - min_y) * 0.2 # Add 10% padding
plt.ylim(min_y - padding, max_y + padding)
# Place the legend outside the plot
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')

# Set ticks for each day
plt.xticks(x, rotation=0)

plt.grid(True)
plt.show()
```
